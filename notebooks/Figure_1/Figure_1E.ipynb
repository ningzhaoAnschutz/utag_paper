## Importing Libraries
___

```python
# Import necessary libraries
import os
from os import listdir
from os.path import join
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import scipy.stats as stats
```

## Directories
___

```python
# List of directory containing data 
current_dir = Path.cwd()
print("Current working directory:", current_dir)
data_dir = current_dir.joinpath('raw_data', 'Figure_1E')
print("Data directory:", data_dir)
plots_dir = current_dir.joinpath('plots', 'Figure_1E')
plots_dir.mkdir(parents=True, exist_ok=True)
print("Plots directory:", plots_dir)
# List all files in the data directory
files = os.listdir(data_dir)
intensity_list = [file for file in files if file.endswith('.xlsx')]
print(intensity_list)
```

```python
# Loading the data
cells = dict()
intensity_data = []
count = 0
for file in intensity_list:
    file_path = os.path.join(data_dir, file)
    cells[count] = pd.read_excel(file_path)
    intensity_data.append(str(file_path)[-17:-5])
    count += 1
print("The number of cells in the folder is:",len(cells), ". Cell data is shown below.")
cells[0] 
```

## Plotting
____

```python
x = cells[0]['Distance_(µm)']
y1 = cells[0]['Normalized Green']
y2 = cells[0]['Normalized Red']
fig = plt.figure(figsize=(2.7, 2.5))
plt.plot(x, y1, 'g', x, y2, 'm', label = intensity_data[0], linewidth=2)
#plt.xlabel("Distance_(µm)")
#plt.ylabel("Normalized Intensity (a.u.)")
#plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1.0))
plt.tick_params(axis='both', which='major', labelsize=14)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
plt.tight_layout()
plt.yticks([0, 0.5, 1], ['0', '0.5', '1'])
plt.xticks([0, 10, 20], ['0', '10', '20'])
box_line = 1
plt.gca().spines['top'].set_linewidth(box_line)
plt.gca().spines['right'].set_linewidth(box_line)
plt.gca().spines['left'].set_linewidth(box_line)
plt.gca().spines['bottom'].set_linewidth(box_line)
plt.savefig(plots_dir.joinpath('Intensity plot for wtscFV-GFP.png'), dpi=300, bbox_inches='tight', transparent=True)
plt.show()
correlation, p_value = stats.pearsonr(y1, y2)
print("Correlation coefficient:", correlation)
print("P-value:", p_value)
```
